```python
dataset = pd.read_excel("D:\Subject\Year 3\DO_AN_2\dataset\preprocessed_laptop.xlsx", engine="openpyxl")
dataset.head(5)
```

```python
def classify_cpu(cpu_name):
    cpu_name = str(cpu_name).lower()

    #tier 0: Thấp
    if any(x in cpu_name for x in ['i3','ryzen 3','core ultra 5', 'xeon']) or any(x in cpu_name for x in ['intel','g1']):
        return 0
    
    #tier 1: Trung
    if any(x in cpu_name for x in ['i7', 'ryzen 7', 'i5', 'ryzen 5', 'core ultra 5', 'core ultra 7']) and any(x in cpu_name for x in ['u', 'p', 'h', 'g7','m']):
        return 1
    if 'core ultra' in cpu_name and ('5' in cpu_name or '7' in cpu_name):
        return 1
    if any(x in cpu_name for x in ['core 5', 'core 7', 'core i7', 'ultra 5','ultra 7']):
        return 1
    
    #tier 2: Cao
    if any(x in cpu_name for x in ['i9', 'ryzen 9', 'i7-13', 'i7-12', 'ryzen 7 7', 'ryzen 7 8', 'i7-11', 'i7-10']) and any(x in cpu_name for x in ['h', 'hx']):
        return 2
    if any(x in cpu_name for x in ['core ultra 7', 'core ultra 9', 'ryzen ai']) and 'h' in cpu_name:
        return 2
    if any(x in cpu_name for x in ['ultra 9']):
        return 2
```

```python
dataset['cpu_tier'] = dataset['CPU'].apply(classify_cpu)
```

```python
def classify_gpu(gpu_name):
    gpu_name = str(gpu_name).lower()

    # Tier 2: GPU rời cao cấp
    if any(x in gpu_name for x in ['rtx 3060', 'rtx 3070', 'rtx 4060', 'rtx 4070', 'rx 6700', 'rx 6800', 'rx 7600']):
        return 2

    # Tier 1: GPU rời cơ bản
    if any(x in gpu_name for x in ['gtx', 'rtx 2050', 'rtx 3050', 'rx 5500', 'rx 5600', 'rx 6500', 'rx 6550', 'rx 6600']):
        return 1

    # Tier 0: GPU tích hợp
    if any(x in gpu_name for x in ['intel', 'amd graphics', 'uhd', 'iris', 'arc']):
        return 0

    return 1
```

```python
dataset['gpu_tier'] = dataset['Gpu'].apply(classify_gpu)
```

```python
dataset[['Gpu','gpu_tier']].head()
```

```python
condition_map = {'New': 3, 'Like New': 2, 'Outlet': 1, 'Used': 0}
dataset['condition_tier'] = dataset['Condition'].map(condition_map)
dataset.head()
```

```python
# Tách độ phân giải (FHD, QHD, 4K...)
dataset['screen_res'] = dataset['Screen'].str.extract(r'(FHD|HD|QHD|4K|UHD|2K|2.5K|3K|3.5K)', expand=False)
```

```python
res_map = {'HD': 0,'FHD': 1,'2K': 2,'2.5K':3,'3K':4,'3.5K':5,'4K':6}
dataset['screen_tier'] = dataset['screen_res'].map(res_map)
```

```python
condition_map = {'Windows 11': 3, 'Windows 10': 2, 'Windows 11 Home': 1, 'Windows': 0}
dataset['os_tier'] = dataset['Os_sys'].map(condition_map)
```

```python
scaler = MinMaxScaler()
dataset[['RAM', 'ROM', 'Weight']] = scaler.fit_transform(dataset[['RAM', 'ROM', 'Weight']])
```

```python
dataset['Prices'] = dataset['Prices']/1000000
```

```python
dataset.head()
```

```python
train_data = dataset[dataset['Prices'].notnull()]
```

```python
test_data = dataset[dataset['Prices'].isna()]
```

```python
x_train = train_data[['Weight','cpu_tier','gpu_tier','condition_tier','screen_tier','os_tier',"RAM",'ROM','Weight']]
y_train = train_data['Prices']
x_test = test_data[['Weight','cpu_tier','gpu_tier','condition_tier','screen_tier','os_tier','RAM','ROM','Weight']]
y_test = test_data['Prices']
```

```python
from sklearn.model_selection import cross_val_score, KFold
from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor, GradientBoostingRegressor
```

```python
models = {
    'Linear Regression': LinearRegression(),
    'Decision Tree': DecisionTreeRegressor(),
    'Random Forest': RandomForestRegressor(),
    'Gradient Boosting': GradientBoostingRegressor()
}
```

```python
results = []
names = []
for name, model in models.items():
  kfold = KFold(n_splits=5, random_state=42, shuffle=True)
  cv_results = cross_val_score(model, x_train, y_train, cv=kfold, scoring='neg_mean_squared_error')
  results.append(cv_results)
  names.append(name)
  msg = "%s: %f (%f)" % (name, cv_results.mean(), cv_results.std())
  print(msg)
```

```python
model = GradientBoostingRegressor(random_state=42)
model.fit(x_train, y_train)
```

```python
y_pred = model.predict(x_test)
print(y_pred)  
```

```python
test_data['Prices'] = y_pred
test_data.head()
```
